```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv('lenta-ru-news.csv',delimiter=',',low_memory=False)
df
```

```python
df = df.dropna(subset=['title','text','tags'])
```

```python
# Convert to list
text_data = df.text.values.tolist()
title_data = df.title.values.tolist()
# Remove Emails
text_data = [re.sub(r'\S*@\S*\s?', '', sent) for sent in text_data]
title_data = [re.sub(r'\S*@\S*\s?', '', sent) for sent in title_data]
# Remove new line characters
text_data = [re.sub(r'\s+', ' ', sent) for sent in text_data]
title_data = [re.sub(r'\S*@\S*\s?', '', sent) for sent in title_data]
# Remove distracting single quotes
text_data = [re.sub(r"\'", "", sent) for sent in text_data]
title_data = [re.sub(r'\S*@\S*\s?', '', sent) for sent in title_data]
text_data[:1]
```
